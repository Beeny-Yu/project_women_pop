```python
import pandas as pd
```

## 성별, 장르별 나눠보기

```python
#여남나눠보기
df_m = pd.read_csv('man_lyrics.csv')
df_w = pd.read_csv('woman_lyrics.csv')
```

```python
df_m['Gender'] = '남자'
df_w['Gender'] = '여자'
```

```python
df_merge = [df_m,df_w]
df_all = pd.concat(df_merge)
df_all
```

```python
example_list = ['입술','hot','피부','body','다리','야해','sexy','몸매',
                '침대','섹시','벗어','booty','허리','엉덩이','벗은','흥분','골반']

test = '|'.join(example_list)

result = df_all[df_all['Lyric'].str.contains(test)]
result
```

```python
#성별&장르별
result_all = result.groupby(['Gender','Genre']).size().reset_index(name='Count')
result_all
```

```python
#성별&장르별&연도별
result_all = result.groupby(['Year','Gender','Genre']).size().reset_index(name='Count')
result_all
```

```python
#Only 장르
pd.pivot_table(result, values='Title', index = 'Genre', aggfunc='count', margins=True)
```
